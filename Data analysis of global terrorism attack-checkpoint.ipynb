```python
##################################IRAQ########################
Iraq=data[data['country']=='Iraq']
```

```python
sns.set_context(context='notebook',font_scale=1.5)
plt.figure(figsize=(16,9))
v1=Iraq['year'].value_counts().to_frame().reset_index().rename(columns={'index':'year','year':'attack'}).sort_values(by='year')
sns.barplot(data=v1,x='year',y='attack',ci=None)
plt.title("HISTORY OF TERRORIST ATTACKS IN IRAQ")
plt.xticks(rotation=90)
```

1. THERE WERE FEW TERRORISM ACTIVITIES IN IRAQ RESPECTIVELY IN THE YEAR 2012 AND 2013.
2. WE CAN SEE ALMOST MORE THAN 2 FOLD RISE IN ATTACKS IN THE YEAR 2014 AND 5 FOLD RISE IN ATTACKS IN THE YEAR 2015.
3. THE YEAR 2016 WITNESSED THE HIGHEST TERRORIST ATTACKS IN IRAQ.'

```python
Iraq[Iraq['group'].isin(Iraq['group'].value_counts()[1:5].index)]\
[['casualties','group']].groupby('group').sum().reset_index()\
.merge(Iraq[Iraq['group'].isin(Iraq['group'].value_counts()[1:5].index)]\
       ['group'].value_counts().to_frame().reset_index().\
       rename(columns={'index':'group','group':'attack'}), on='group').\
sort_values(by='attack',ascending=False).plot.barh(x='group',y=['attack','casualties'],\
                                                   figsize=(12,5),stacked=True,width=0.5)
plt.title("MOST ACTIVE TERRORIST GROUPS IN IRAQ")
```

1. ISIL IS THE MOST ACTIVE GROUP IN IRAQ WITH HIGHEST NUMBER OF TERRORIST ATTACKS AND CASUALTIES FOLLOWED BY AI-QAIDA.

```python
Iraq[Iraq['city'].isin(Iraq['city'].value_counts()[0:10].index)]['city'].\
value_counts().to_frame().plot.bar(figsize=(16,6),width=0.3)
plt.ylabel('ATTACK')
plt.xlabel('CITIES')
plt.title("MOST AFFECTED CITIES IN IRAQ")
```

1. BAGHDAD WITNESSED MORE THAN 5 FOLD NUMBER OF ATTACKS THAN THE REST CITIES.

```python
v1=Iraq[Iraq['group'].isin(Iraq['group'].value_counts()[1:5].index)]
pd.crosstab(v1['group'],v1['attack']).plot.barh(stacked=True,figsize=(9,5),width=0.5)
plt.legend(loc=9,bbox_to_anchor=(1.32,1))
plt.title("ATTACK TYPE")
```

1. ISIL, AI-QAIDA USED BOMBING/EXPLOSION AS THEIR PRIMARY ATTACK TYPE.
2. THE GROUP ASAIB AHL AL_HAQQ ADOPTED FACILITY/INFRASTRUCTURE ATTACK.

```python
pd.crosstab(v1['group'],v1['weapon_type']).plot.barh(stacked=True,figsize=(9,5),width=0.5)
plt.legend(loc=9,bbox_to_anchor=(0.8,0.75))
plt.title("WEAPONS USED BY DIFFERENT GROUPS")
```

1. ISIL,AI-QAIDA USED EXPLOSIVES MAINLY TO SPREAD ITS TERROR.
2. THE GROUP ASAIB AHL AL_HAQQ USED INCENDIARY.

```python
v1=v1[v1['target'].isin(v1['target'].value_counts()[0:10].index)]
pd.crosstab(v1['group'],v1['target']).plot.barh(stacked=True,figsize=(9,5),width=0.7)
plt.legend(loc=9,bbox_to_anchor=(1.27,1))
plt.title("FAVORITE TARGETS OF DIFFERENT TERRORIST GROUPS")
```

1. PRIVATE CITIZENS & PROPERTY , MILITARY ARE THE MAJOR TARGETS OF ISIL.

```python
list1=np.unique(data["year"])
list2=list1.tolist()
a=np.zeros(len(list2))
j=0
for i in list2:
    a[j]=data[(data.country=="Iraq")&(data.year==i)][["casualties"]].sum()
    j+=1
slices=a.tolist()
cols=['magenta','cyan','brown','gold','blue']
plt.pie(slices,labels=list2,colors=cols,startangle=0,explode=(0.1,0.3,0,0,0.1),shadow=True,autopct='%.1f%%')
plt.title('CASUALTIES PER YEAR IN IRAQ')
plt.legend(loc='best', bbox_to_anchor=(0.5, 1.0, 1.5, 0.5))
plt.show()
```

1. MAXIMUM CASUALTY IN IRAQ OCCURED IN THE YEAR 2016.
2. CUMULATIVE CASUALTY FOR THE YEAR 2012,2013 AND 2014 IS APPROXIMATELY EQUAL TO THAT OF 2015.
